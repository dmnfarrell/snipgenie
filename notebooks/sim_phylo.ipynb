# Create M.bovis phylogeny from outbreak simulation

```r
library(ape)
library(phangorn)
library('TransPhylo')
set.seed(0)
#library('TreeDist')
setwd('/storage/btbgenie/simpop/')
```

```r
neg=100/365
off.r=3
w.shape=10
w.scale=0.1
pi=0.25
simu <- simulateOutbreak(neg=neg,pi=pi,off.r=off.r,w.shape=w.shape,
                         w.scale=w.scale,dateStartOutbreak=2008,dateT=2012)
plot(simu)
```

```r
ttree<-extractTTree(simu)
plot(ttree)
ptree<-extractPTree(simu)
plot(ptree)
p<-phyloFromPTree(ptree)
plot(p)
axisPhylo(backward = F)
#write.tree(p,'sim.newick')
```

## add 'clusters' to tree

```r
c<-child(p, '2')
plot(c)
```
